# **1. Prepare Data for Training**

*   data_loader using *torchvision.datasets.ImageFolder* for Custom dataset
*   **image augmentation** in *transforms*
*   include augmented face image samples in report



```python
# import packages
import torch
import torchvision
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision import datasets
```

```python
# transforms for image augmentation
train_transform = torchvision.transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.3,), (1.0,)),
    transforms.Resize(112),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(45),
])

val_transform = torchvision.transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.3,), (1.0,)),
    transforms.Resize(112),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(45),
])
```

```python
train_path = '/content/drive/MyDrive/인공지능개론 프로젝트/train'
val_path = '/content/drive/MyDrive/인공지능개론 프로젝트/val'

# write ImageFolder code below
train_data = datasets.ImageFolder(train_path, train_transform)
val_data = datasets.ImageFolder(val_path, val_transform)

# check the label
train_data.class_to_idx
val_data.class_to_idx
```

# **2. Prepare Model**

*   Pytorch ResNet - *ref*. https://github.com/pytorch/vision/blob/main/torchvision/models/resnet.py
*   use **ResNet50** from torchvision.model_zoo
*   explore more models in https://pytorch.org/vision/stable/models.html
*   **change the dimension of the classifier**

```python
# assign device cpu or gpu
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
# import packages for importing models
import torchvision.models as models
import torch.nn as nn
import torch.nn.functional as F
```

```python
# load model and change the # of classes
model = torchvision.models.resnet50(pretrained=True, progress=True)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 1)
model = model.to(device)
```

# **3. Training**


*   write **training code** including belows:
   - hyper parameters such as batch size, learning rate, epoch
   - criterion(loss function such as BCELoss), optimizer(eg. Adam, SGD, etc.)  and scheduler
   - save model weight

*   **print training/validation loss and accuracy** per epoch or iteration
*   inlcude visualizer, **tensorboard**, to show training/validation accuracy and loss


```python
import torch
import wandb
import sklearn
from sklearn import metrics
import datetime
import os
import torch.optim as optim
import numpy as np
from tqdm import tqdm

wandb.login()
```

```python
save_path = "/content/drive/MyDrive/인공지능개론 프로젝트/Result/" +  datetime.datetime.now().strftime("%Y%m%d-%H%M%S") + "/"
os.makedirs(save_path, exist_ok=True)
```

```python
# training and validation code can be writed in one function. It's your taste!
```

```python
# Hyperparameters
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

config  = {
    'epochs': 50,
    'classes': 2,
    'batch_size': 64,
    'learning_rate': 1e-2,
    'architecture': 'resnet50'
}

wandb.init(project="MASK_KangLim", config=config, name='MASK_KangLim')
```

```python
# optimizer, loss, scheduler
optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
loss_fn = nn.BCEWithLogitsLoss()
scheduler = optim.lr_scheduler.LambdaLR(optimizer=optimizer, lr_lambda=lambda epoch: 0.95 ** epoch)

min_loss = np.inf
```

```python
# data_loader
batch_size = 64

train_loader = DataLoader(dataset=train_data,
                         batch_size=batch_size,
                         shuffle=True)

val_loader = DataLoader(dataset=val_data,
                         batch_size=batch_size,
                         shuffle=True)
```

```python
# ROC curve, AUC (Hint: use sklearn or wandb function, using sklearn to extract fpr, tpr will be bonus score)
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Array initialization for ROC curve
train_labels = []
train_predictions = []
val_labels = []
val_predictions = []
```

```python
# Training and validation

for epoch in range(config['epochs']):

  # train_model
  model.train()
  running_loss = 0
  acc_temp = 0

  for img, tlbl in tqdm(train_loader):
      img, tlbl = img.to(device), tlbl.to(device)
      tlbl = tlbl.unsqueeze(1)
      batch_size = img.size(0)
      optimizer.zero_grad()
      output = model(img)
      loss = loss_fn(output.float(), tlbl.float())
      loss.backward()
      optimizer.step()

      tpred = output
      tpred[tpred >= 0.5] = 1
      tpred[tpred < 0.5] = 0
      acc_temp += tpred.eq(tlbl).int().sum()
      running_loss += loss.item() * batch_size

      train_labels.extend(tlbl.tolist())
      train_predictions.extend(output.squeeze().tolist())
  scheduler.step()

  train_acc = acc_temp / len(train_loader.dataset)
  train_loss = running_loss / len(train_loader.dataset)

  # val_model
  model.eval()
  running_loss_val = 0
  acc_temp_val = 0

  for img, vlbl in val_loader:
      img, vlbl = img.to(device), vlbl.to(device)
      vlbl = vlbl.unsqueeze(1)
      batch_size = img.size(0)
      output = model(img)

      vpred = output
      vpred[vpred >= 0.5] = 1
      vpred[vpred < 0.5] = 0
      acc_temp_val += vpred.eq(vlbl).int().sum()
      running_loss_val += loss_fn(output.float(), vlbl.float()).item() * batch_size

      val_labels.extend(vlbl.tolist())
      val_predictions.extend(output.squeeze().tolist())

  val_acc = acc_temp_val / len(val_loader.dataset)
  val_loss = running_loss_val / len(val_loader.dataset)

  train_labels = np.array(train_labels)
  train_predictions = np.array(train_predictions)
  val_labels = np.array(val_labels)
  val_predictions = np.array(val_predictions)

  # Extraction of FPR, TPR from train, val
  fpr_train, tpr_train, thresholds_train = roc_curve(train_labels, train_predictions)
  roc_auc_train = auc(fpr_train, tpr_train)
  fpr_val, tpr_val, thresholds_val = roc_curve(val_labels, val_predictions)
  roc_auc_val = auc(fpr_val, tpr_val)

  train_labels = train_labels.tolist()
  train_predictions = train_predictions.tolist()
  val_labels = val_labels.tolist()
  val_predictions = val_predictions.tolist()

  wandb.log({
    "train_loss": train_loss,
    "val_loss": val_loss,
    "train_acc": train_acc,
    "val_acc": val_acc,
    "Train AUC": roc_auc_train,
    "Val AUC": roc_auc_val
  })

  if val_loss < min_loss:
      print(f'val_loss improved from {min_loss:.5f} to {val_loss:.5f}. Model saved.')
      min_loss = val_loss
      torch.save(model.state_dict(), save_path + 'resnet50.pth')

  print(f'epoch {epoch+1:02d}, loss: {train_loss:.5f}, acc: {train_acc:.5f}, val_loss: {val_loss:.5f}, val_acc: {val_acc:.5f}, train_auc: {roc_auc_train:.5f}, val_auc: {roc_auc_val:.5f}')

# ROC curve
plt.figure()
plt.plot(fpr_train, tpr_train, label='Train ROC curve (area = %0.2f)' % roc_auc_train)
plt.plot(fpr_val, tpr_val, label='Validation ROC curve (area = %0.2f)' % roc_auc_val)
plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()


wandb.finish()
```
